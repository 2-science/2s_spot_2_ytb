```python
def collecting_spotify_files(files_directory, extension, verbose=True):
    
    print(f"Looking at dir: {files_directory}")
    print(f"Looking for files with extension: .{extension} \n")
    
    
    ## Collect all files with specific extension
    all_files_at_dir = glob.glob(files_directory+f"\\*.{extension}")

    Nfiles = len(all_files_at_dir)
    all_names_at_dir = []

    if verbose:
        print(f"Found {Nfiles} files! These are: \n")
    for i in range(Nfiles):
        temp_name = all_files_at_dir[i].replace(files_directory+'\\', "")
        all_names_at_dir.append(temp_name.replace(f".{extension}", "")) # ".jpg"))
        print(all_files_at_dir[i])

    return all_files_at_dir, all_names_at_dir
```

## Transforming audio to video

```python
import ffmpeg
import moviepy.editor as mp
from moviepy.editor import ImageClip, concatenate_videoclips
```

```python
def audio_2_video(audio_file, image_file, logo_image, fps=5):
    
    video_file = audio_file.replace(".m4a", ".mp4")
    
    # 1. Create an audio clip from the M4A file
    audio_clip = mp.AudioFileClip(audio_file)

    # 2. Create a video clip from the image file
    # Set the duration of the image clip to match the audio clip's duration
    image_clip = ImageClip(image_file).set_duration(audio_clip.duration-5)
    logo_clip = ImageClip(logo_image).set_duration(5)

    final_clip = concatenate_videoclips([image_clip, logo_clip])  ##, method="compose"
    
    # 3. Set the audio of the video clip to the loaded audio
    video_clip = final_clip.set_audio(audio_clip)

    ## Make horizontal aspect ratio, to not upload as short
    video_clip = video_clip.resize(newsize=(1920, 1080))
    
    # 4. Write the final video file (MP4 format)
    # YouTube recommends the H.264 video codec and AAC audio codec
    video_clip.write_videofile(
        video_file,
        codec="libx264",         # Video codec (H.264)
        audio_codec="aac",       # Audio codec (AAC)
        temp_audiofile='temp-audio.m4a', # Temporary file for audio processing
        remove_temp=True,        # Remove the temporary audio file
        logger=None,              # Suppress unnecessary logs if preferred
        fps=8
    )
    
    # Close clips to free up resources
    audio_clip.close()
    image_clip.close()
    logo_clip.close()
    video_clip.close()
    
    print(f"Successfully converted {audio_file} to {video_file} for YouTube. \n")
```

```python
def automate_process(audio_flist, audio_fnames, images_flist, images_fnames, video_fnames, logo_image, fps=5):

    ## NOTE: The file_list need to be more than 2 files
    ## to work properly (needs to be a list).
    
    for idx, name in enumerate(images_fnames):
        if (name in audio_fnames) and (name not in video_fnames):
            print(f"I will transform the file with name: {name}")
            audio_name = os.getcwd()+'\\Video_Audio_files\\'+name+'.m4a'
            image_name = os.getcwd()+'\\Articles_images\\'+name+'.jpg'

            audio_2_video(audio_name, image_name, logo_image, fps=fps)

    print()
    print("All files transformed!")
```
